```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
x = winequality.drop('quality',axis=1)
x
```

```python
winequality.replace({'quality':{3:0,4:0,5:0,6:0,7:1,8:1}},inplace = True)
```

```python
winequality['quality'].value_counts()
```

```python
y = winequality['quality']
y
```

```python
from imblearn.over_sampling import SMOTE
#SYNTHETIC MINORITY OVERSAMPLING TECHNIQUES
```

```python
smote=SMOTE()
x_resampled,y_resampled = smote.fit_resample(x,y)
```

```python
x_resampled.shape
```

```python
y_resampled.value_counts()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_resampled,y_resampled,test_size=0.3,random_state=42)
```

```python
x_resampled.shape,x_train.shape,x_test.shape,y_resampled.shape,y_train.shape,y_train.shape
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
train_data_predict = model.predict(x_train)
train_data_accuracy = accuracy_score(train_data_predict,y_train)
print('Accuracy of trainingData',train_data_accuracy)
```

```python
test_data_predict = model.predict(x_test)
test_data_accuracy = accuracy_score(test_data_predict,y_test)
print('Accuracy of Testingdata',test_data_accuracy)
```

```python
cm = confusion_matrix(test_data_predict,y_test)
cm
```

```python
sns.heatmap(cm,annot=True,annot_kws ={'size':20},fmt ='.1f',cmap = 'coolwarm',square = True,xticklabels = ['predicted_true','predicted_false'],yticklabels = ['actual_true','actual_false'])
plt.xlabel('predicted_area')
plt.ylabel('actual_data')
plt.title('condusion_matix')
plt.show()
```

```python
print(classification_report(test_data_predict,y_test))
```

```python
new_data = (7.4,0.700,0.00,1.9,0.076,11.0,34.0,0.99780,3.51,0.56,9.4)
new_data_as_np = np.asarray(new_data)
new_data_reshaped = new_data_as_np.reshape(1,-1)
prediction = model.predict(new_data_reshaped)
if prediction[0]==0:
    print('low quality')
else:
    print('high quality')
```
